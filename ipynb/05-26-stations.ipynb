```python
import matplotlib.pyplot as plt
# from mpl_toolkits.basemap import Basemap
import numpy as np
```

```python
lines = station_data.strip().split('\n')
lons, lats, names = [], [], []

for line in lines:
    parts = line.split()
    lon = float(parts[0])
    lat = float(parts[1])
    name = ' '.join(parts[3:])
    lons.append(lon)
    lats.append(lat)
    names.append(name)
```

```python
from sithom.plot import plot_defaults
plot_defaults()
fig, axs = plt.subplots(1, 1, figsize=(20, 20))
plt.scatter(lons, lats, marker='o', color='red', zorder=5)
for lon, lat, name in zip(lons, lats, names):
    plt.text(lon, lat, name, fontsize=9, ha='right', va='top')

plt.xlabel('Longitude [$^\circ$E]')
plt.ylabel('Latitude [$^\circ$N]')
plt.title('Station Locations')
plt.show()
```

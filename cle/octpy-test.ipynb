```python
from sithom.io import read_json
from sithom.time import timeit

@timeit
def run_call(**kwargs):

    in_dict = read_json("data/inputs.json")
    in_dict.update(kwargs)
    octave.push("in", in_dict)
    octave.eval("""[rr, VV, rmax, rmerge, Vmerge] = ...
    ER11E04_nondim_r0input(in.Vmax, in.r0, in.fcor, in.Cdvary, in.Cd, in.w_cool, in.CkCdvary, in.CkCd, in.eye_adj, in.alpha_eye);""")
    return octave.pull("out")
```

```python
run_call()
```
